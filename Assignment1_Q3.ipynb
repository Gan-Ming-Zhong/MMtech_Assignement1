```python
from moviepy.editor import *
vid = VideoFileClip('/content/drive/MyDrive/1_Code/ECE3086/Assignment1/surveillance_5.mp4')
start_time = float(input('Enter Start Time')) * 60
end_time = float(input('Enter End Time')) * 60
trim_vid = vid.subclip(start_time,end_time)
trim_vid.write_videofile('/content/drive/MyDrive/1_Code/ECE3086/Assignment1/10minsurveillance_5.mp4', codec = 'libx264')
```

```python
#Read the video
cam = cv2.VideoCapture("10minsurveillance_5.mp4")

try:
  
  if not os.path.exists('data'):
    os.makedirs('data')

except OSError:
  print("Error: Creating Directory of Data")

currentframe = 0

while(True):

  ret,frame = cam.read()

  if ret:

    name = './data/frame' + str(currentframe) + ".jpg"
    print("Creating..." + name)
    cv2.imwrite(name,frame)
    currentframe +=1

  else:

    break

cam.release()
```

```python
video = cv2.VideoCapture("surveillance_5.mp4");
fps = video.get(cv2.CAP_PROP_FPS)
print(format(fps))
height = video.get(cv2.CAP_PROP_FRAME_HEIGHT)
print(height)
width = video.get(cv2.CAP_PROP_FRAME_WIDTH)
print(width)
video.release()
```
